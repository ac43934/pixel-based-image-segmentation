```python
import pyelastix
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline
# To read the image data we use imageio
import imageio
# Read image data
im1 = imageio.imread('C:/test/1.png')
im2 = imageio.imread('C:/test/2.png')
#im2 = imageio.imread('https://dl.dropboxusercontent.com/u/1463853/images/chelsea_morph1.png')

# Select one channel (grayscale), and make float
im2 = im2[:,:,1].astype('float32')
im1 = im1[:,:,1].astype('float32')

# Get default params and adjust
params = pyelastix.get_default_params(type='AFFINE')
params.MaximumNumberOfIterations = 500
params.NumberOfResolutions = 10
#params.FinalGridSpacingInVoxels = 10
print(params)

#ELASTIX_PATH = 'C:/Users/69654/AppData/Local/Continuum/Anaconda2/Lib/pyelastix.pyc/'
#exe = pyelastix.get_elastix_exes()
# Register!
im3, field = pyelastix.register(im1, im2, params)

# Visualize the result
fig = plt.figure(figsize=(50,50));
#plt.clf()
#plt.subplot(231); plt.imshow(im1)
#plt.subplot(232); plt.imshow(im2)
plt.subplot(231); plt.imshow(im3)
#plt.subplot(232); plt.imshow(field[0])
#plt.subplot(233); plt.imshow(field[1])
plt.imsave("C:/segmented/AFFINE_38_position_4.png", im3)
#if hasattr(plt, 'use'):
 #   plt.use().Run()  # visvis
#else:
 #   plt.show()
```
