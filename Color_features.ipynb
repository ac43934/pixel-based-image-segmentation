```python
from skimage.io import imread, imsave
from PIL import Image
import matplotlib as plt
import numpy as np
from skimage.io import imshow, show
from skimage import io, color
def color_features(path):
    image = Image.open(path)
    X = np.reshape(image, [-1, 3])
    original_shape = X.shape
    #GRAY = color.rgb2gray(X)
    lab = color.rgb2lab(image)
    hsv = color.rgb2hsv(image)
    yiq = color.rgb2yiq(image)
    xyz = color.rgb2xyz(image)
    rgbcie = color.rgb2rgbcie(image)
    hed = color.rgb2hed(image)
    ypbpr = color.rgb2ypbpr(image)
    ycbcr = color.rgb2ycbcr(image)
    yuv = color.rgb2yuv(image)
    RGB= X
    LAB=lab.reshape(-1,3)
    YUV= yuv.reshape(-1,3)
    HSV= hsv.reshape(-1,3)
    YIQ= yiq.reshape(-1,3)
    XYZ= xyz.reshape(-1,3)
    RGBCIE= rgbcie.reshape(-1,3)
    HED= hed.reshape(-1,3)
    YPBPR= ypbpr.reshape(-1,3)
    YCBCR= ycbcr.reshape(-1,3)
    S = np.column_stack((RGB,HSV,LAB,YUV,YIQ,XYZ,RGBCIE,HED,YPBPR,YCBCR))
    #Smax, Smin = S.max(), S.min()
    #Z = (S - Smin)/(Smax - Smin)
    return (S)
```
